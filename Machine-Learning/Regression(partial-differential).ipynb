# 회귀(Regression)
    편미분 방정식의 결과만을 사용하여 회귀 계수를 구하는 방법(경사하강법을 사용하지 않음)

```python
import numpy as np 
import matplotlib.pyplot as plt 

np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 6 + 4 * X + np.random.randn(100, 1)

plt.scatter(X, y)
```

## 손실 함수(Loss Function)인 MSE(Mean Sqaure Error)를 계수인 w1과 w0에 대해 각각 편미분하여 이를 0(최솟값)이 되게 하는 편미분 방정식을 w1과 w0에 대해 나타냄

```python
def partial_differential(w0, w1, X, y):
    w1 = (np.mean(X * y) - np.mean(X) * np.mean(y)) / (np.mean(np.square(X)) - np.square(np.mean(X))) # 계수 w1의 편미분 방정식 결과
    w0 = np.mean(y) - w1 * np.mean(X) # 계수 w0의 편미분 방정식 결과
    
    return w1, w0

w0 = 0
w1 = 0
w1, w0 = partial_differential(w0, w1, X, y)
print(w1, w0)
```

```python
y_pred = w1 * X + w0
plt.scatter(X, y)
plt.plot(X, y_pred, c='red')
```
